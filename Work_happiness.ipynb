```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
correlation_matrix = df_numerized.corr(method = 'pearson' )
plt.rcParams["figure.figsize"] = (32,15) #(width x Height)
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for All Features')
plt.xlabel('Employee Attributes')
plt.ylabel('Employee Attributes')
plt.show()
```

Based on the above correlation matrix, Job Level, Monthly Income, and Total Working Years have a high correlation coefficient of aropund 0.77 - 0.78.
